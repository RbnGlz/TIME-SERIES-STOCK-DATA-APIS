```python
import yfinance as yf

# Definir el ticker del futuro del S&P 500
ticker = "ES=F"

# Descargar los datos históricos (últimos 6 meses por defecto)
data = yf.download(ticker, period="1y", interval="1d")  # Cambia el periodo si necesitas más datos

# Mostrar los primeros datos
print(data.head())
```

```python
# Graficar el precio de cierre para visualizar la coherencia de la serie temporal

import matplotlib.pyplot as plt
data['Close'].plot(figsize=(12, 6), title="Futuros S&P 500 (ES=F)")
plt.xlabel("Fecha")
plt.ylabel("Precio de Cierre")
plt.show()
```

```python
# Guardar en un archivo CSV
filename = "sp500_futuros.csv"
data.to_csv(filename)

print(f"Archivo guardado como: {filename}")
```

```python

# Invertir el orden de las filas para tener la serie temporal en orden cronológico

import pandas as pd

# Paso 1: Leer el archivo CSV
archivo_csv = "sp500_futuros.csv"  # Asegúrate de que el archivo esté en el mismo directorio o proporciona la ruta completa
df = pd.read_csv(archivo_csv)

# Paso 2: Invertir el orden de las filas
df_invertido = df.iloc[::-1].reset_index(drop=True)  # reset_index(drop=True) reinicia los índices después de invertir

# Paso 3: Guardar el archivo modificado
archivo_modificado = "sp500_futuros_invertido.csv"  # Nombre del nuevo archivo
df_invertido.to_csv(archivo_modificado, index=False)

print(f"El archivo ha sido invertido y guardado como '{archivo_modificado}'.")
```
